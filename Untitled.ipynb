```python
import matplotlib.pyplot as plt
import os

os.environ['PROJ_LIB'] = r'c:\Users\ALI\anaconda3\Library\share'
from mpl_toolkits.basemap import Basemap
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# Function to calculate ground track
def calculate_ground_track(inclination, periods, time_step):
    earth_radius = 6371  # Earth's radius in kilometers
    altitude = 500  # Satellite altitude in kilometers
    orbital_period = 2 * np.pi * np.sqrt((earth_radius + altitude)**3 / 398600.4418)  # Kepler's third law
    num_steps = int(periods * orbital_period / time_step)
    
    latitudes = []
    longitudes = []
    
    for step in range(num_steps):
        time = step * time_step
        mean_anomaly = 2 * np.pi * (time % orbital_period) / orbital_period
        true_anomaly = mean_anomaly  # Simplified assumption of circular orbit
        
        lat = np.degrees(np.arcsin(np.sin(np.radians(inclination)) * np.sin(true_anomaly)))
        lon = np.degrees(true_anomaly)
        
        latitudes.append(lat)
        longitudes.append(lon % 360 - 180)
    
    return latitudes, longitudes

# Set up the map
fig, ax = plt.subplots(figsize=(14, 7))
m = Basemap(projection='cyl', resolution='c', ax=ax)

m.drawcoastlines()
m.drawcountries()
m.drawparallels(np.arange(-90., 91., 30.), labels=[1, 0, 0, 0])
m.drawmeridians(np.arange(-180., 181., 60.), labels=[0, 0, 0, 1])

# Plot ground tracks for different inclinations
inclinations = [43,53, 70, 97.6]
colors = ['red', 'blue', 'green']
labels = ['43 inc.','53 inc.', '70 inc.', '97.6 inc.']

for inclination, color, label in zip(inclinations, colors, labels):
    latitudes, longitudes = calculate_ground_track(inclination, periods=3, time_step=60)
    m.plot(longitudes, latitudes, color=color, label=label)

# Plot legend
plt.legend()

# Title and show plot
plt.title('Satellite Ground Tracks with Different Inclinations')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import datetime as dt
import math as m

# Constants
mu = 398600.4418  # Earth's gravitational parameter [km^3/s^2]
RE = 6378  # Earth's radius [km]
altitude = 550  # Typical altitude for Starlink satellites [km]
r = RE + altitude
D = 24 * 0.997269  # Earth's sidereal day [hours]

def plot_orbits(inclinations):
    fig = plt.figure()
    ax = plt.axes(projection='3d', computed_zorder=False)
    
    for inc in inclinations:
        a = r  # Semi-major axis for circular orbit
        e = 0  # Eccentricity for circular orbit
        i = np.radians(inc)  # Inclination in radians
        RAAN = np.radians(0)  # Right Ascension of Ascending Node in radians
        omega = np.radians(0)  # Argument of perigee in radians

        orb = {"name": f"{inc} Inclination", "a": a, "e": e, "i": i, "RAAN": RAAN, "omega": omega}
        orb.update({"b": orb["a"] * m.sqrt(1 - orb["e"]**2), "c": orb["a"] * orb["e"]})
        
        R = np.matmul(np.array([[m.cos(orb["RAAN"]), -m.sin(orb["RAAN"]), 0],
                                [m.sin(orb["RAAN"]), m.cos(orb["RAAN"]), 0],
                                [0, 0, 1]]),
                      (np.array([[1, 0, 0],
                                 [0, m.cos(orb["i"]), -m.sin(orb["i"])],
                                 [0, m.sin(orb["i"]), m.cos(orb["i"])]])))
        R = np.matmul(R, np.array([[m.cos(orb["omega"]), -m.sin(orb["omega"]), 0],
                                   [m.sin(orb["omega"]), m.cos(orb["omega"]), 0],
                                   [0, 0, 1]]))
        
        x, y, z = [], [], []
        for theta in np.linspace(0, 2 * m.pi, 100):
            P = np.matmul(R, np.array([[orb["a"] * m.cos(theta)],
                                       [orb["b"] * m.sin(theta)],
                                       [0]])) - np.matmul(R, np.array([[orb["c"]], [0], [0]]))
            x.append(P[0, 0])
            y.append(P[1, 0])
            z.append(P[2, 0])
        
        ax.plot(x, y, z, zorder=5, label=orb["name"])
    
    u, v = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]
    ax.plot_wireframe(RE * np.cos(u) * np.sin(v), RE * np.sin(u) * np.sin(v), RE * np.cos(v), color="b", alpha=0.5, lw=0.5, zorder=0)
    plt.title("Orbits plotted in the ECI frame")
    ax.set_xlabel("X-axis (km)")
    ax.set_ylabel("Y-axis (km)")
    ax.set_zlabel("Z-axis (km)")
    ax.xaxis.set_tick_params(labelsize=7)
    ax.yaxis.set_tick_params(labelsize=7)
    ax.zaxis.set_tick_params(labelsize=7)
    ax.set_aspect('auto')
    ax.legend()
    plt.show()

# List of inclinations
inclinations = [43, 53, 70, 96]

# Plot the orbits
plot_orbits(inclinations)
```
